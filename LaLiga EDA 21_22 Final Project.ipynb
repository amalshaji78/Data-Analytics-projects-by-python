```python
#total goals

total_goals = data['goals'].sum()
total_goals
```

```python
#total penalty goals

total_penalty_goals = data['penalty_goals'].sum()
total_penalty_goals
```

```python
#total penalty concided

total_penalties_conceded = data['penalties_conceded'].sum()
total_penalties_conceded
```

```python
#create pie-chart for penalty scored and missed

plt.figure(figsize=(13,6))
plt_not_scrd = data['penalties_conceded'].sum()-total_penalty_goals
data1 = [plt_not_scrd,total_penalty_goals]
labels = ['Penalities missed','Penalities Scored']
color = sns.color_palette('pastel')
plt.pie(data1,labels = labels , colors = color, autopct = '%.0f%%')
plt.show()

```

```python
#total goals from outside box

total_goal_from_outside_box = data['goals_from_outside_box'].sum()
total_goal_from_outside_box
```

```python
#total_goals_from_inside_box

total_goal_from_inside_box = data['goals_from_inside_box'].sum()
total_goal_from_inside_box
```

```python
#create piechart for goals from outside and inside box

plt.figure(figsize=(13,6))
data2 = [total_goal_from_outside_box,total_goal_from_inside_box]
labels = ['Outside Box','Inside Box']
color = sns.color_palette('pastel')
plt.pie(data2,labels = labels , colors = color, autopct = '%.0f%%')

plt.show()
```

```python
#total goals from right foot

total_right_foot_goal = data['right_foot_goals'].sum()
total_right_foot_goal
```

```python
#create pie chart for different types of goals


plt.figure(figsize=(15,8))
data3 = [total_right_foot_goal,total_left_foot_goals,total_goal_header ]
labels = ['Right Foot','Left Foot','Header']
color = sns.color_palette('pastel')
plt.pie(data3,labels = labels , colors = color, autopct = '%.0f%%')


plt.show()
```
