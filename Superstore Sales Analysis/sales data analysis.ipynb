```python
#getting month year from the dataset

sales['month_year'] = sales['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales['month_year'] 
```

```python
sales.groupby('month_year').sum()
```

```python
sales_monthly_trend = sales.groupby('month_year').sum()['sales'].reset_index()
sales_monthly_trend
```

```python
#graph of monthy sales

plt.figure(figsize=(15,6))
plt.plot(sales_monthly_trend['month_year'],sales_monthly_trend['sales'])
plt.xticks(rotation=90,size=8)
plt.show()
```

```python
#top 10 products by sales

top_sales = sales.groupby('product_name')['sales'].sum().sort_values(ascending=False)
top_sales.head(10)
```

```python
#most selling products

most_selling = sales.groupby('product_name')['quantity'].sum().sort_values(ascending=False)
most_selling.head(10)
```

```python
#most preffered shipmode

sns.countplot(x='ship_mode',data = sales)
```

```python
#most profitable category and sub category

profitable_category = pd.DataFrame(sales.groupby(['category','sub_category'])['profit'].sum()).sort_values(['category','profit'], ascending=False)
profitable_category
```

```python
#find some corelations 

sales.corr()
```

```python
correlation_matrix = sales.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
#top countries in sales
top_countries = sales.groupby('country')['sales'].sum().sort_values(ascending=False)
top_countries.head(20)
```

```python
#top countries in profit
top_countries_profit = sales.groupby('country')['profit'].sum().sort_values(ascending=False)
top_countries_profit.head(20)
```

```python
#top countries in discount
top_countries_discount = sales.groupby('country')['discount'].sum().sort_values(ascending=False)
top_countries_discount.head(20)
```
